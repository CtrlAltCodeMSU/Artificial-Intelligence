```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)   # 100 points from 0 to 10 
y1 = np.sin(x)                # Sine function 
y2 = np.cos(x)                # Cosine function 
y3 = np.random.rand(10)       # Random values for bar plot 
categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Create a figure and a set of subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Line Plot
axs[0].plot(x, y1, label='Sine', color='blue')
axs[0].plot(x, y2, label='Cosine', color='orange')
axs[0].set_title('Line Plot of Sine and Cosine Functions')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].legend()
axs[0].grid()

# Bar Plot
axs[1].bar(categories, y3, color='green')
axs[1].set_title('Bar Plot of Random Values')
axs[1].set_xlabel('Categories')
axs[1].set_ylabel('Values')
axs[1].grid(axis='y')

# Scatter Plot
axs[2].scatter(x, y1, color='purple', label='Sine Data Points', alpha=0.6)
axs[2].scatter(x, y2, color='red', label='Cosine Data Points', alpha=0.6)
axs[2].set_title('Scatter Plot of Sine and Cosine Functions')
axs[2].set_xlabel('X-axis')
axs[2].set_ylabel('Y-axis')
axs[2].legend()
axs[2].grid()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
